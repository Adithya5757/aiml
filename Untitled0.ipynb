```python
!pip install opencv-python face_recognition matplotlib numpy
import cv2
import face_recognition
import matplotlib.pyplot as plt
import numpy as np

# Load a test image (you can upload your own later)
!wget https://raw.githubusercontent.com/ageitgey/face_recognition/master/examples/obama.jpg

# Read and display the image
image = face_recognition.load_image_file("obama.jpg")
face_locations = face_recognition.face_locations(image)

# Draw box around face
for (top, right, bottom, left) in face_locations:
    cv2.rectangle(image, (left, top), (right, bottom), (0, 255, 0), 2)

# Show image
plt.imshow(image)
plt.axis('off')
plt.title("Face Detection in Action")
plt.show()
```
